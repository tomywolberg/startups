```python
def plot_outliers(df):
    num_cols = df.select_dtypes(include=['float64', 'int64']).shape[1]
    num_rows = (num_cols - 1) // 2 + 1
    fig, axs = plt.subplots(num_rows, 2, figsize=(20, 5*num_rows))
    axs = axs.flatten()
    for i, col in enumerate(df.select_dtypes(include=['float64', 'int64']).columns):
        box = axs[i].boxplot(df[col], patch_artist=True, boxprops=dict(facecolor='#336fa2'), medianprops=dict(color='black'))
        for patch in box['fliers']:
            patch.set_markerfacecolor('black')
            patch.set_markeredgecolor('black')
        axs[i].set_title(col)
    plt.tight_layout()
    plt.show()
```

```python
plot_outliers(cons)
```

```python
cons = cons[cons['Admin_Costs'] > 45000]
cons = cons[cons['Net_Profit'] > 0]
cons = cons[cons['Profit'] > 0]
```

```python
cons.shape
```

```python
plot_outliers(cons)
```

```python
plt.figure(figsize = (10, 10))

sns.heatmap(cons.select_dtypes(include=['float64', 'int64']).corr(), cmap = 'coolwarm', annot = True, annot_kws = {"size": 10})
```

```python
import statsmodels.formula.api as smf
```

```python
x = cons.select_dtypes(include=['float64', 'int64']).drop(['Net_Profit', 'Profit'], axis = 1).columns
```

```python
formula = "Net_Profit ~ " + ' + '.join(x)
```

```python
model = smf.ols(formula = formula, data = cons).fit()
print(model.summary())
```

```python
predictedNP = model.predict(cons[x])
cons['residuals'] = cons.Net_Profit - predictedNP
cons.head()
```

```python
cons['residuals'].mean()
```

```python
cons['EC'] = cons.residuals**2
```

```python
cons.head()
```

```python
cons.EC.mean()
```

```python
industria = cons.groupby('Industry_Sector').agg(Net_Profit = ('Net_Profit', 'mean')).reset_index()
industria
```
